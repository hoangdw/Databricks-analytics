```sql
-- Total sales
SELECT SUM(sales_amount) AS total_sales FROM workspace.databricks_analytics.fact_sales;

-- Total customers
SELECT COUNT(DISTINCT customer_key) AS total_customers FROM workspace.databricks_analytics.dim_customers;

-- Total products
SELECT COUNT(DISTINCT product_key) AS total_products FROM workspace.databricks_analytics.dim_products;

-- Total orders
SELECT COUNT(DISTINCT order_number) AS total_orders FROM workspace.databricks_analytics.fact_sales;

-- Sales by product
SELECT product_key, SUM(sales_amount) AS sales_by_product FROM workspace.databricks_analytics.fact_sales GROUP BY product_key;

-- Sales by category
SELECT p.category, SUM(f.sales_amount) AS sales_by_category
FROM workspace.databricks_analytics.fact_sales f
JOIN workspace.databricks_analytics.dim_products p ON f.product_key = p.product_key
GROUP BY p.category;

-- Orders over time
SELECT order_date, COUNT(DISTINCT order_number) AS orders_count
FROM workspace.databricks_analytics.fact_sales
GROUP BY order_date
ORDER BY order_date;
```
